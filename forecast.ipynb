```python
#predict tomorrow's max temp using today's data
core_weather["target"] = core_weather.shift(-1)["max_temp"]
core_weather = core_weather.iloc[:-1,:].copy()
core_weather 
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["precip", "snow", "snow_depth", "max_temp", "min_temp"]
```

```python
# split data into train and test sets
train = core_weather.loc[:"2021-12-31"]
test = core_weather.loc["2022-01-01":]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
predictions
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "prediction"]
combined
```

```python
combined.plot()
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2021-12-31"]
    test = core_weather.loc["2022-01-01":]

    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])

    error = mean_squared_error(test["target"], predictions)
    
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```
